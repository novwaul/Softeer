```python
print('trans:', data['am'].nunique())
print('gear:', data['gear'].nunique())
```

```python
cnt = pd.crosstab(data['gear'], data['am'])
print(cnt)
```

```python
summary = dict()
for i in data['gear'].unique().tolist():
    for j in data['am'].unique().tolist():
        if (i,j) not in summary:
            summary[(i,j)] = 0

for a,b in zip(data['gear'].tolist(), data['am'].tolist()):
    summary[(a, b)] += 1

summary
```

```python
xs = list()
ys = list()
for k,v in summary.items():
    strk = '(' + str(k[0]) + ', ' + str(k[1]) + ')'
    xs.append(strk)
    ys.append(v)

plt.bar(xs,ys)
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.show()
```

```python
row = 4
maxcol = 3
cnt = 1
plt.suptitle('Histograms of Variables')

for name in data.columns:
    if name != 'car name':
        plt.subplot(row, maxcol, cnt)
        plt.hist(data[name])
        plt.title(name)
        plt.grid()
        cnt += 1
plt.tight_layout()
plt.show()
```
